```python
# -*- coding: utf-8 -*-

# Define o texto do prompt inicial para instruir o Agente
primeiro_prompt = """Voc√™ √© um Agente de Design de Interiores e Planejamento de Espa√ßos... (mantenha o texto inicial e as instru√ß√µes de BRIEFING)

**AP√ìS O BRIEFING E QUANDO O CLIENTE CONFIRMAR QUE QUER SUGEST√ïES:**
... (mantenha as instru√ß√µes sobre a fase de sugest√µes e refer√™ncias) ...
- Pergunte se as sugest√µes foram √∫teis e se o cliente gostaria de explorar outra √°rea (como otimiza√ß√£o de espa√ßo ou d√∫vidas sobre planta).

**QUANDO O CLIENTE PEDIR AJUDA ESPEC√çFICA COM OTIMIZA√á√ÉO DE ESPA√áO OU PLANEJADOS (EX: CLOSET, ARM√ÅRIO, CANTO, NICHO):**
- Reconhe√ßa imediatamente que o cliente mudou de t√≥pico para otimiza√ß√£o de espa√ßo.
- Mostre entusiasmo em ajudar com essa parte pr√°tica.
- **√â CRUCIAL pedir detalhes:** Pe√ßa ao cliente para fornecer informa√ß√µes espec√≠ficas para que suas sugest√µes sejam relevantes. Pe√ßa:
    - O ambiente ou a √°rea exata que ele quer otimizar (ex: "o closet do quarto", "o arm√°rio da cozinha perto da pia", "um canto espec√≠fico na sala").
    - AS MEDIDAS PRECISAS dessa √°rea (largura, profundidade, altura). **Se as medidas n√£o forem fornecidas, pe√ßa-as.**
    - O que ele precisa armazenar ou a fun√ß√£o espec√≠fica daquele espa√ßo (ex: "principalmente sapatos e bolsas", "panelas grandes e eletrodom√©sticos", "livros e objetos de decora√ß√£o").
    - Quais problemas atuais ele enfrenta nessa √°rea (ex: "sempre bagun√ßado", "espa√ßo desperdi√ßado no alto", "dif√≠cil encontrar as coisas").
- **Com base NAS MEDIDAS E NECESSIDADES FORNECIDAS pelo cliente, gere SUGEST√ïES CONCRETAS para otimiza√ß√£o:**
    - Ideias de layout interno (ex: sugerir prateleiras ajust√°veis, tipos de gavetas, onde colocar cabideiros para roupas longas, nichos para itens pequenos).
    - Tipos de organizadores que funcionariam bem com aquelas medidas e uso (ex: caixas empilh√°veis, divis√≥rias internas, suportes espec√≠ficos).
    - Sugest√µes de profundidades ideais para prateleiras ou arm√°rios dependendo do que ser√° guardado.
    - Ideias criativas para aproveitar cantos, espa√ßos estreitos ou √°reas de dif√≠cil acesso.
- **Sempre reforce que estas s√£o sugest√µes baseadas nas medidas fornecidas** e que um profissional de marcenaria ou design pode criar uma solu√ß√£o sob medida.
- Pergunte se as sugest√µes de otimiza√ß√£o foram √∫teis ou se o cliente tem mais perguntas sobre outras √°reas do projeto.

**QUANDO O CLIENTE PEDIR AJUDA COM A PLANTA DO IM√ìVEL:**
- Reconhe√ßa que o cliente quer falar sobre a planta ou o desenho do im√≥vel.
- **√â FUNDAMENTAL explicar a sua limita√ß√£o:** Diga educadamente e de forma clara que, como Agente de IA neste formato de chat de texto, voc√™ n√£o pode "ver" a imagem ou o PDF da planta diretamente.
- Diga que, no entanto, voc√™ pode ajudar MUITO se ele DESCREVER o que ele est√° vendo na planta ou se ele tiver uma pergunta espec√≠fica baseada no texto, n√∫meros ou s√≠mbolos que ele identificou no desenho.
- **Pe√ßa explicitamente para o cliente DESCREVER o ponto da planta que gera a d√∫vida para que voc√™ possa tentar interpretar:**
    - Qual ambiente ou c√¥modo ele est√° olhando na planta?
    - O que ele v√™ ou l√™ exatamente naquele ponto que est√° gerando a d√∫vida (ex: "vejo um s√≠mbolo com um 'X' no meio da sala", "tem uma linha com o n√∫mero '2.55' em cima", "o que significa essa √°rea hachurada?", "tem uma legenda que n√£o entendi")?
    - Qual √© a d√∫vida espec√≠fica sobre essa descri√ß√£o (ex: "O que este s√≠mbolo representa?", "Esta medida √© de qual elemento?", "Esta parede √© estrutural?", "O que est√° escrito nesta legenda?")
- **Com base APENAS na descri√ß√£o detalhada do cliente e no seu conhecimento geral** sobre simbologia comum de plantas de arquitetura, conven√ß√µes de desenho t√©cnico e termos de constru√ß√£o civil (conhecimento que o modelo Gemini possui), tente responder √† pergunta da forma mais √∫til e educativa poss√≠vel.
- **SEJA SEMPRE CAUTELOSO E N√ÉO AFIRME COM CERTEZA ABSOLUTA:** Como voc√™ n√£o "v√™" a planta, suas respostas s√£o interpreta√ß√µes baseadas em descri√ß√µes textuais e conhecimento geral. Sempre reforce que a interpreta√ß√£o mais precisa e a informa√ß√£o oficial est√£o na planta original do arquiteto/engenheiro ou ao consultar o profissional respons√°vel. Use frases como:
    - "Com base na sua descri√ß√£o de 's√≠mbolo com X na sala', isso GERALMENTE representa um ponto de el√©trica (como teto ou parede) ou de l√≥gica, mas confirme na legenda da sua planta."
    - "Pelo que voc√™ descreveu ('linha com o n√∫mero 2.55 em cima'), parece ser a medida (2 metros e 55 cent√≠metros) de um elemento como uma parede, porta ou janela naquela se√ß√£o, mas verifique o que a linha indica na sua planta e a unidade usada (metros ou cent√≠metros)."
    - "N√£o posso confirmar se uma parede √© estrutural apenas pela descri√ß√£o. Plantas estruturais mostram isso, ou um engenheiro/arquiteto pode confirmar no local."
    - "O que est√° escrito 'xxx' na legenda pode significar 'yyy', mas a legenda da SUA planta √© que define o significado exato."
- Mantenha um tom de ajuda, paci√™ncia e colabora√ß√£o, guiando o cliente a descrever para que voc√™ possa ajudar.
- Ao final da resposta sobre a planta, pergunte se a explica√ß√£o foi √∫til e se ele tem outras d√∫vidas sobre a planta ou quer voltar a falar de design ou otimiza√ß√£o.

**FORMATO DA RESPOSTA:**
- Use par√°grafos para separar ideias.
- Use bullet points (`- ` ou `* ` ) para listas, se apropriado (como nas se√ß√µes de sugest√µes ou otimiza√ß√£o).
- **NOVA INSTRU√á√ÉO:** **Quando apresentar listas de op√ß√µes (ex: tipos de pisos), caracter√≠sticas de elementos, pr√≥s/contras, ou fizer compara√ß√µes que envolvam m√∫ltiplos atributos, use a formata√ß√£o de **tabela Markdown** (usando `|` para colunas e `-` para linhas de separa√ß√£o) para organizar a informa√ß√£o de forma clara e estruturada.** Exemplo de formato de tabela:
  `| Cabe√ßalho 1 | Cabe√ßalho 2 |`
  `|---|---|`
  `| Dado A1 | Dado A2 |`
  `| Dado B1 | Dado B2 |`
- Inclua quebras de linha para facilitar a leitura.
- N√£o entregue um bloco √∫nico de texto.
- Mantenha o tom amig√°vel, profissional e √∫til.

Comece se apresentando como seu Agente de Design e fazendo a primeira pergunta para iniciar o briefing, focando em qual ambiente ou c√¥modo o cliente gostaria de focar primeiro.
"""

print("--- INICIANDO AGENTE ---")

try:
    print("AGENTE (Pensando na primeira pergunta...):")
    response = chat.send_message(primeiro_prompt)

    # Imprime a primeira resposta do Agente
    agente_primeira_resposta = response.text
    print("AGENTE:", agente_primeira_resposta)
    print("\n--- COME√áO DO BRIEFING ---")

    # Loop para intera√ß√£o cont√≠nua com o usu√°rio
    while True:
        # Pede a entrada do usu√°rio
        user_input = input("Aluri: ")

        # Condi√ß√£o para sair da conversa
        if user_input.lower() == 'sair':
            print("AGENTE: Ok, encerrando a sess√£o de briefing por agora. Se precisar, √© s√≥ chamar!")
            break # Sai do loop

        # Envia a resposta do usu√°rio para o chat (mantendo o hist√≥rico)
        # O modelo usa o hist√≥rico e o prompt inicial para gerar a pr√≥xima resposta
        print("AGENTE (Processando...):")
        response = chat.send_message(user_input)

        # Imprime a resposta do Agente
        print("AGENTE:", response.text)
        print("-" * 20) # Separador para organizar a conversa

except Exception as e:
    print(f"\nERRO DURANTE A INTERA√á√ÉO: {e}")
    print("Verifique sua conex√£o, API Key ou se houve algum problema com a resposta do modelo.")

print("--- FIM DA SESS√ÉO ---")
```

```python
# -*- coding: utf-8 -*-

# --- 1. Importar Bibliotecas ---
# ipywidgets geralmente j√° vem pr√©-instalado no Google Colab.
# Certifique-se que voc√™ executou a c√©lula ANTERIOR que instala a biblioteca google-generativeai.
import ipywidgets as widgets
from IPython.display import display, clear_output # Fun√ß√µes para mostrar widgets e limpar sa√≠da

# --- 2. Verificar e Usar o Agente Inicializado ---
# √â FUNDAMENTAL que voc√™ j√° tenha rodado as c√©lulas ANTERIORES que fazem:
# a) Configura√ß√£o da API Key com genai.configure(api_key=...)
# b) Inicializa√ß√£o do Modelo Gemini: model = genai.GenerativeModel(...)
# c) Inicializa√ß√£o da Sess√£o de Chat: chat = model.start_chat(history=[])
# d) Envio do SEU PROMPT INICIAL COMPLETO para o chat: chat.send_message(seu_prompt_completo)
# O objeto 'chat' j√° deve existir e estar configurado quando esta c√©lula rodar.


# --- 3. Definir a Fun√ß√£o que Processa as Mensagens ---
# Esta fun√ß√£o pega o input do usu√°rio da interface, envia para o Agente (chat),
# e retorna a resposta do Agente.
def process_user_message_for_ipywidgets(user_input):
    # Se o input for vazio ou s√≥ espa√ßos, n√£o processa
    if not user_input or user_input.strip() == "":
        return None # Retorna None para indicar que nada deve ser impresso como resposta do Agente

    try:
        # Opcional: Adiciona um pequeno delay para simular "pensamento"
        # import time
        # time.sleep(1.5)

        # Envia a mensagem do usu√°rio para a sess√£o de chat do Gemini
        # O objeto 'chat' gerencia o hist√≥rico da conversa automaticamente
        response = chat.send_message(user_input)

        # Retorna o texto da resposta do Agente
        # As formata√ß√µes (par√°grafos, listas) que voc√™ pediu no prompt
        # vir√£o aqui e o print exibir√°.
        return response.text

    except Exception as e:
        # Em caso de erro na comunica√ß√£o, imprime no output_area e retorna uma mensagem de erro
        with output_area: # Garante que esta mensagem de erro v√° para a √°rea correta
             print(f"\n[ERRO na Comunica√ß√£o]: Ocorreu um problema ao falar com o Agente. Detalhes: {e}\n")
        return "Ops! Parece que houve um erro. Poderia tentar enviar sua mensagem novamente?"


# --- 4. Construir os Componentes Visuais da Interface ---
# Cria os elementos interativos: √°rea para mostrar a conversa, caixa para digitar, bot√£o para enviar.
output_area = widgets.Output() # √Årea onde a conversa ser√° exibida. Print's direcionados para c√°.

input_box = widgets.Text( # Caixa de texto para o usu√°rio digitar
    placeholder='Digite sua mensagem aqui...',
    description='Voc√™:', # R√≥tulo ao lado da caixa de texto
    disabled=False,
    layout=widgets.Layout(width='80%') # Define a largura para ocupar 80% do espa√ßo horizontal dispon√≠vel
)

send_button = widgets.Button( # Bot√£o para enviar a mensagem
    description='Enviar',
    disabled=False,
    button_style='success', # Bot√£o verde para indicar a√ß√£o principal
    tooltip='Clique para enviar sua mensagem ao Agente', # Texto que aparece quando o mouse passa por cima
    icon='paper-plane', # √çcone (requer que o Colab carregue os √≠cones Font Awesome)
    layout=widgets.Layout(width='20%') # Define a largura para ocupar 20% do espa√ßo horizontal
)


# --- 5. Definir o que acontece quando o Bot√£o √© Clicado ---
# Esta fun√ß√£o √© chamada toda vez que o bot√£o 'Enviar' √© clicado.
def on_send_button_click(b):
    # 1. Pega o texto que o usu√°rio digitou na caixa de input
    user_message = input_box.value

    # 2. Limpa a caixa de input imediatamente para que o usu√°rio possa digitar a pr√≥xima mensagem
    input_box.value = ''

    # 3. Usa a √°rea de output para exibir esta intera√ß√£o
    with output_area:
        # Opcional: clear_output(wait=True) aqui limparia TODA a conversa anterior no output_area,
        # mostrando apenas a √∫ltima intera√ß√£o. Normalmente, queremos ver o hist√≥rico, ent√£o deixamos comentado.

        # 4. Exibe a mensagem que o usu√°rio acabou de enviar, com um r√≥tulo claro
        print(f"VOC√ä: {user_message}")

        # 5. Verifica se o usu√°rio digitou o comando de sa√≠da ANTES de chamar a IA
        if user_message.lower() == 'sair':
             print("AGENTE: Ok, encerrando a sess√£o de chat. Se precisar, √© s√≥ chamar!")
             # Desabilita a caixa de input e o bot√£o para encerrar a intera√ß√£o
             input_box.disabled = True
             send_button.disabled = True
             return # Sai da fun√ß√£o, n√£o chama a IA se o comando for 'sair'

        # 6. Processa a mensagem do usu√°rio com o Agente (chama a fun√ß√£o que interage com o chat)
        agent_response_text = process_user_message_for_ipywidgets(user_input=user_message)

        # 7. Exibe a resposta do Agente na √°rea de output, S√ì SE HOUVE RESPOSTA
        if agent_response_text is not None:
            print(f"AGENTE: {agent_response_text}")

        # 8. Adiciona um separador visual para organizar a conversa
        print("-" * 60) # Linha pontilhada/tracejada como separador


# --- 6. Vincular o Bot√£o √† Fun√ß√£o ---
# Quando o evento 'click' acontecer no send_button, chame a fun√ß√£o on_send_button_click.
send_button.on_click(on_send_button_click)


# --- 7. Exibir os Componentes Visuais na Sa√≠da da C√©lula ---
# Mostra a √°rea onde as mensagens aparecer√£o.
display(output_area)
# Mostra a caixa de input e o bot√£o lado a lado, abaixo da √°rea de output.
display(widgets.HBox([input_box, send_button]))


# --- 8. Exibir a SUA MENSAGEM de Apresenta√ß√£o e Instru√ß√µes INICIAIS no Output ---
# Este bloco executa uma vez quando a c√©lula √© rodada.
# Ele imprime o texto que voc√™ pediu DENTRO da √°rea de output, no in√≠cio da conversa.
with output_area:
    # Opcional: clear_output() aqui limparia QUALQUER output anterior desta √°rea ao re-executar a c√©lula.
    # clear_output()

    # Define O TEXTO ESPEC√çFICO QUE VOC√ä QUER como mensagem inicial
    initial_welcome_and_instructions = """
üè° Agente de Design de Interiores e Planejamento com IA

Bem-vindo ao seu Agente de Design virtual!

Este notebook cont√©m um agente de conversa√ß√£o interativo constru√≠do com a API do Google Gemini, focado em auxiliar no processo de design de interiores e planejamento de espa√ßos.

**Como Usar:**

* Certifique-se que as c√©lulas de configura√ß√£o da API e inicializa√ß√£o do Agente foram executadas acima.
* Execute esta c√©lula de c√≥digo para carregar a interface interativa.
* A interface aparecer√° na sa√≠da desta c√©lula. Digite suas mensagens na caixa de texto e clique em 'Enviar'.
* O Agente responder√° na √°rea de output acima.
* Digite 'sair' para encerrar a conversa.

O Agente est√° pronto para te ajudar com briefing, sugest√µes de design, otimiza√ß√£o de espa√ßo e d√∫vidas sobre a planta! :)
"""
    # Imprime este texto na √°rea de output. Ele aparecer√° primeiro na conversa.
    print(initial_welcome_and_instructions)
    print("-" * 60) # Adiciona um separador visual ap√≥s a mensagem inicial


# Mensagem no log do Colab (vis√≠vel abaixo da sa√≠da da c√©lula, n√£o na √°rea de output)
print("\n--- Interface do Agente com ipywidgets Carregada Abaixo na Sa√≠da da C√©lula ---")
```
